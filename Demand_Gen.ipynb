```python
#importing libraries

import pandas as pd
import numpy as np
```

```python
#loading data and storing as dataframes

qlick = pd.read_excel('Qlik_over_under_mtd_all_Markets_all_Nameplates (1).xlsx')
demand = pd.read_excel('Demand Objectives Aug_20230817_ unhide - Mfundo.xlsx', sheet_name= 'Raw')
qlick.head()
```

```python
# Function to lowercase text and search for specific words
def search_and_store(text):
    text = text.lower()
    words_to_search = ['fiesta', 'focus', 'kuga', 'mach-e', 'puma', 'ranger', 'tourneo connect', 'tourneo custom', 'transit', 'transit connect', 'transit courier', 'transit custom']  # Replace with your specific words
    found_words = [word for word in words_to_search if word in text]
    return ', '.join(found_words)

# Apply the function to the 'Model' and store results in a new 'Nameplate' column
qlick['Nameplate'] = qlick['Model'].apply(search_and_store)
qlick['Nameplate'].dropna()
qlick['Nameplate'].reset_index(drop=True, inplace=True)
qlick['Nameplate'].unique()
```

```python
#Filtering for the available nameplates

qlick = qlick[(qlick['Nameplate'] == 'focus')|(qlick['Nameplate'] == 'fiesta')|(qlick['Nameplate'] == 'kuga')|(qlick['Nameplate'] == 'mach-e')|(qlick['Nameplate'] == 'tourneo connect')]
```

```python
#Filtering for the columns I will use

sub_qlick = qlick[["Market", "Nameplate", "O/(U) Cal'd Obj"]]
sub_qlick['Market'] = sub_qlick['Market'].str.lower() #lower cases to match demand dataframe
sub_qlick
```

```python
#Filtering for the columns I will use and lower the nameplates

sub_demand = demand[['MARKET', 'NAMEPLATE', 'Forecast Visits', 'Forecast B&P','Forecast Leads','Visit objectives','B&P objectives', 'Lead objectives']]
sub_demand['NAMEPLATE'] = sub_demand['NAMEPLATE'].str.lower()
sub_demand
```

```python
#lower cases to match the dataframes

sub_qlick.columns = sub_qlick.columns.str.lower()
sub_demand.columns = sub_demand.columns.str.lower()
```

```python
#Merging the two dataframes

result = pd.merge(sub_demand, sub_qlick, how="right", on=["market", "nameplate"])
result['market'].unique()
```

```python
#Filtering to the EU5 countries

eu5 = result[(result['market'] == 'britain')|(result['market'] == 'france')|(result['market'] == 'italy')|(result['market'] == 'spain')|(result['market'] == 'germany')]
eu5.reset_index(drop=True, inplace=True)
eu5
```

```python
#filling all nulls to 0's and replace -'s with 0's

eu5 = eu5.fillna(0)
eu5 = eu5.replace('-', 0)
eu5
```

```python
#Percentage calculation between Forcasts and Objectives with ERROR HANDLING

eu5['% of visits'] = eu5.apply(lambda x: x['forecast visits']/x['visit objectives'] if x['visit objectives'] != 0 else 0, axis=1)
eu5['% of b&p'] = eu5.apply(lambda x: x['forecast b&p']/x['b&p objectives'] if x['b&p objectives'] != 0 else 0, axis=1)
eu5['% of Leads'] = eu5.apply(lambda x: x['forecast leads']/x['lead objectives'] if x['lead objectives'] != 0 else 0, axis=1)
```

```python
eu5.info()
```

```python
# Function to convert floats to integers
def convert_float_to_int(value):
    if isinstance(value, float):
        return int(value)
    else:
        return value

# Apply the conversion function to each cell in the DataFrame
eu5 = eu5.applymap(convert_float_to_int)
```

```python
eu5["o/(u) cal'd obj"] = eu5["o/(u) cal'd obj"].astype(float)
eu5["o/(u) cal'd obj"] = eu5["o/(u) cal'd obj"].astype(int)
```

```python
# Function to calculate estimates with 5% variance

def estimate_b(row):
    # Calculate a random number with 5% variance
    variance = 0.05 * row['% of visits']
    estimate = np.random.uniform(row['% of visits'] - variance, row['% of visits'] + variance)
    return estimate

# Apply the function to each row and create a new column 'Estimated_B'

eu5['Estimated_B'] = eu5["o/(u) cal'd obj"].apply(estimate_b)

# Print the DataFrame with estimated values
print(eu5)
```
